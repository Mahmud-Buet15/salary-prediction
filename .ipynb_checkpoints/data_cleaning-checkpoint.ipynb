## Cleaning Salary

```python
#removing records which don't have salary

# df[df['Salary Estimate']=='-1']   #showing all records with N/A salary

df =df[df['Salary Estimate']!='-1']
```

```python
df.head(30)
```

```python
df['Salary Estimate']
```

```python
#Steps to follow

#get first part of the record (contains salary info)
#replace 'K' and '$'

#create new columns with minimum and maximum range of a salary
#create average
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('(')[0])   #get first part of the record (contains salary info)
```

```python
minus_Kd=salary.apply(lambda x: x.replace('$','').replace('K',''))     #replace 'K' and '$'
```

```python
minus_Kd.head(50)  #showing top 50
minus_Kd.tail(50)  #showing bottom 50

print(minus_Kd.to_string())   #showing full data
```

```python
#  print(minus_Kd.apply(lambda x: 1 if 'Employer Provided Salary:'.lower() in x.lower() else 0 ).to_string())

df['employer_provided']=minus_Kd.apply(lambda x: 1 if 'Employer Provided Salary:'.lower() in x.lower() else 0 )    #create new columns with employer provided info
df.head()
```

```python
df['hourly']=minus_Kd.apply(lambda x: 1 if 'Per Hour'.lower() in x.lower() else 0 )  #create new columns with houlrly info
df.head()
```

```python
print(df[['hourly','employer_provided','Salary Estimate']].to_string())
```

```python
# print(minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:','')).to_string())

min_max=minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
#create new columns with minimum and maximum range of a salary
df['min_salary']=min_max.apply(lambda x: int(x.split('-')[0]))
df['max_salary']=min_max.apply(lambda x: int(x.split('-')[1]))
```

```python
df.head()
df['min_salary'].dtype
```

```python
df['avg_salary']=(df['min_salary']+df['max_salary'])/2
```

```python
df[['min_salary','max_salary','avg_salary']].describe()
```

```python
df.head()
```

## Cleaning Company Name

```python
print(df[['Company Name','Rating']].to_string())
```

```python
df[df['Rating']<0]

# print(df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1).to_string())

df['company_txt']=df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
```

```python
df.head()
```

## Cleaning Location

```python
df['state']=df['Location'].apply(lambda x: x.split(',')[-1])
```

```python
df.state.value_counts()
```

```python
df.head()

# need to check if the job location is at the head quarter

print(df['Headquarters'].apply(lambda x: x.split(',')[-1] ).to_string())
```

```python
print(df[['Headquarters','Location']].to_string())
```

```python
df['same_state']=df.apply(lambda x: 1 if x['Headquarters']==x['Location'] else 0,axis=1)
```

```python
df.same_state.value_counts()
```
